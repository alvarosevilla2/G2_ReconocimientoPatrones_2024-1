## <center>Acerca de los datos </center>

​**Este conjunto de datos contiene 2126 registros de características extraídas de exámenes de Cardiotocograma, que luego fueron clasificados por un obstetra experto en 3 clases: "Normal", "Suspect" y "Pathological". El conjunto de datos contiene las siguientes características:**
    
* **baseline value:** Valor de línea base de la frecuencia cardíaca fetal (FHR) (latidos por minuto)   
* **accelerations:** Número de aceleraciones por segundo
* **fetal_movement:** Número de movimientos fetales por segundo    
* **uterine_contractions:** Número de contracciones uterinas por segundo
* **light_decelerations:** Número de desaceleraciones leves (LDs) por segundo
* **severe_decelerations:** Número de desaceleraciones severas (SDs) por segundo
* **prolongued_decelerations:** Número de desaceleraciones prolongadas (PDs) por segundo
* **abnormal_short_term_variability:** Porcentaje de tiempo con variabilidad anormal a corto plazo
* **mean_value_of_short_term_variability:** Valor medio de la variabilidad a corto plazo
* **percentage_of_time_with_abnormal_long_term_variability:** Porcentaje de tiempo con variabilidad anormal a largo plazo
* **mean_value_of_long_term_variability:** Valor medio de la variabilidad a largo plazo
* **histogram_width:** Ancho del histograma creado usando todos los valores de un registro
* **histogram_min:** Valor mínimo del histograma
* **histogram_max:** Valor máximo del histograma
* **histogram_number_of_peaks:** Número de picos en el histograma del examen
* **histogram_number_of_zeroes:** Número de ceros en el histograma del examen
* **histogram_mode:** Modo del histograma
* **histogram_mean:** Media del histograma
* **histogram_median:** Mediana del histograma
* **histogram_variance:** Varianza del histograma
* **histogram_tendency:** Tendencia del histograma
* **fetal_health:** Codificado como 1-Normal(Normal); 2-Sospechoso(Suspect); 3-Patológic(Pathological).

Vamos a analizar eliminar los valores atípicos de las columnas menos del target.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Copiamos el dataframe original
df_copiado = df.copy()
df_copiado = df_copiado.drop('fetal_health', axis=1) #es nuestro target
```

```python
# Definimos una función para filtrar los valores atípicos basados en el rango intercuartílico
def filtrar_atipicos(columna):
    q1 = columna.quantile(0.25)
    q3 = columna.quantile(0.75)
    iqr = q3 - q1
    umbral_inferior = q1 - 1.5 * iqr
    umbral_superior = q3 + 1.5 * iqr
    return columna[(columna >= umbral_inferior) & (columna <= umbral_superior)]

# Aplicamos al función a df_copiado
df_copiado_filtrado = df_copiado.apply(filtrar_atipicos)

# Vemos la diferencia entre lo antes de filtrar y lo filtrado
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
df_copiado.boxplot()
plt.title('Distribuciones Originales')
plt.subplot(2, 1, 2)
df_copiado_filtrado.boxplot()
plt.title('Distribuciones Filtradas')
plt.tight_layout()
plt.show()
```

```python
df_copiado_filtrado.info()
```

Podemos ver que se ha eliminado un gran número de datos para la columna histogram_numer_of_zeroes, y al ser un porcentaje considerable (20% del dataset), se optó por eliminar la columna y quedarnos con el dataset original menos dicha columna.

```python
df_copiado = df_copiado.drop('histogram_number_of_zeroes', axis=1)
```

```python
df_copiado.info()
```

Ahora vamos a ver las características que serán tomadas en cuenta para el modelo.

```python
# Calculamos la matriz de correlación entre las características filtradas
correlation_matrix = df_copiado.corr().abs()

# Visualizamos la matriz de correlación usando un mapa de calor
import seaborn as sns
sns.heatmap(correlation_matrix)

# Creamos una matriz con True en los elementos por encima de la diagonal principal y False en los demás
upper_triangle = np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool)

# Seleccionamos solo la mitad superior de la matriz de correlación para evitar duplicados
upper_half = correlation_matrix.where(upper_triangle)

# Identificamos las columnas con una correlación mayor que un umbral dado (0.6 en este caso)
highly_correlated_features = [columna for columna in upper_half.columns if any(upper_half[columna] > 0.6)]

# Eliminamos las características altamente correlacionadas del DataFrame original filtrado
df_selected_features = df_copiado.drop(highly_correlated_features, axis=1)
```

##Balanceo de la data

Hay un desbalanceo de la data, vamos a balancearlo haciendo un oversampling.



```python
# Importamos la librería de SMOTE
from imblearn.over_sampling import SMOTE

# Dividir el DataFrame en características (X) y variable objetivo (y)
X = df.drop('fetal_health', axis=1)  # Features
y = df['fetal_health']  # Target variable

# Aplicar SMOTE para realizar oversampling
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Crear un nuevo DataFrame con los datos balanceados
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled['fetal_health'] = y_resampled
```

```python
df_resampled['fetal_health'].value_counts()
```

## GridSearch CV

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline

# Dividimos el DataFrame en características (X) y variable objetivo (y)
X = df_resampled[df_selected_features.columns] # Columnas calculados anteriormente
y = df_resampled['fetal_health']  # Columna 'fetal_health' como variable objetivo

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0)
pipe = Pipeline([('scaler', StandardScaler()), ('svc', SVC())])
# The pipeline can be used as any other estimator
# and avoids leaking the test set into the train set
pipe.fit(X_train, y_train).score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Grid de parámetros
param_grid = [
  {'svc__C': [1, 10, 100, 1000], 'svc__kernel': ['linear']},
  {'svc__C': [1, 10, 100, 1000], 'svc__gamma': [0.001, 0.0001], 'svc__kernel': ['rbf']},
  {'svc__C': [1, 10, 100, 1000], 'svc__degree': [2, 3, 4], 'svc__kernel': ['poly']},
]


grid_search = GridSearchCV(pipe, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
print("Mejores parámetros:", grid_search.best_params_)
print("Precisión del mejor modelo:", grid_search.best_score_)
y_pred = grid_search.predict(X_test)

# Rendimiento en los datos de prueba
accuracy = accuracy_score(y_test, y_pred)
print("Precisión en datos de prueba:", accuracy)
```
